## Exercise 1

** Follow along with these steps: **
* ** Create a figure object called fig using plt.figure() **
* ** Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax. **
* ** Plot (x,y) on that axes and set the labels and titles to match the plot below:**

```python
import matplotlib.pyplot as plt

# Create a figure object
fig = plt.figure()

# Add an axis to the figure canvas at [0, 0, 1, 1]
ax = fig.add_axes([0, 0, 1, 1])

# Data for plotting
x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]

# Plot (x, y) on the axis
ax.plot(x, y)

# Set labels and title
ax.set_title('Title of the Plot')
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')

# Display the plot
plt.show()
```

## Exercise 3

** Create the plot below by adding two axes to a figure object at [0,0,1,1] and [0.2,0.5,.4,.4]**

```python
import matplotlib.pyplot as plt

# Create a figure object
fig = plt.figure()

# Add the first axis to the figure at [0, 0, 1, 1]
ax1 = fig.add_axes([0, 0, 1, 1])

# Add the second axis to the figure at [0.2, 0.5, 0.4, 0.4]
ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.4])

# Data for plotting
x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]

# Plot (x, y) on the first axis
ax1.plot(x, y, label='Main Plot')
ax1.set_title('Main Plot')
ax1.legend()

# Plot (x, y) on the second axis
ax2.plot(x, y, color='red', label='Inset Plot')
ax2.set_title('Inset Plot')
ax2.legend()

# Show the figure
plt.show()
```

** Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:**

```python
import matplotlib.pyplot as plt

# Create some example data
x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]
z = [0, 1, 8, 27, 64]

# Create a figure object
fig = plt.figure()

# Add the first axis (main plot) at [0, 0, 1, 1]
ax1 = fig.add_axes([0, 0, 1, 1])
ax1.plot(x, y, label='y = x^2')
ax1.plot(x, z, label='z = x^3')
ax1.set_title('Main Plot')
ax1.legend()

# Add the second axis (inset plot) at [0.2, 0.5, 0.4, 0.4]
ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.4])
ax2.plot(x, y, label='y = x^2', color='blue')
ax2.plot(x, z, label='z = x^3', color='red')
ax2.set_title('Inset Plot')

# Set x and y limits for the inset plot
ax2.set_xlim(1, 3)
ax2.set_ylim(0, 30)

# Add legends to the inset plot
ax2.legend()

# Show the figure
plt.show()
```

## Exercise 4

** Use plt.subplots(nrows=1, ncols=2) to create the plot below.**

```python
import matplotlib.pyplot as plt

# Example data
x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]
z = [0, 1, 8, 27, 64]

# Create subplots with 1 row and 2 columns
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))

# First subplot
axes[0].plot(x, y, label='y = x^2', color='blue')
axes[0].set_title('Plot 1')
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].legend()

# Second subplot
axes[1].plot(x, z, label='z = x^3', color='red')
axes[1].set_title('Plot 2')
axes[1].set_xlabel('x')
axes[1].set_ylabel('z')
axes[1].legend()

# Adjust layout and display the plots
plt.tight_layout()
plt.show()
```

** Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**

```python
import matplotlib.pyplot as plt

# Example data
x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]
z = [0, 1, 8, 27, 64]

# Create subplots with 1 row and 2 columns
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

# First subplot: (x, y)
axes[0].plot(x, y, label='y = x^2', color='blue', linewidth=2, linestyle='--')
axes[0].set_title('Plot of y = x^2')
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].legend()

# Second subplot: (x, z)
axes[1].plot(x, z, label='z = x^3', color='green', linewidth=3, linestyle='-.')
axes[1].set_title('Plot of z = x^3')
axes[1].set_xlabel('x')
axes[1].set_ylabel('z')
axes[1].legend()

# Adjust layout and display the plots
plt.tight_layout()
plt.show()
```
